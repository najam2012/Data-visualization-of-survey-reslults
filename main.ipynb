# Correlation

finding correlation between variables using pearson method

```python
CM=result.corr(method ='pearson') 
CM
```

```python
import matplotlib.pyplot as plt
f = plt.figure(figsize=(19, 15))
plt.matshow(CM.corr(), fignum=f.number)
plt.xticks(range(CM.shape[1]), CM.columns, fontsize=14, rotation=45)
plt.yticks(range(CM.shape[1]), CM.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16);
```
